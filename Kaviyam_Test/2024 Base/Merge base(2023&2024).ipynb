```python
import pandas as pd

# Load the two datasets
base_2023 = pd.read_excel("Pvt Car Jan'23 to  Dec 23 base Final.xlsx")
base_2024 = pd.read_excel("cleaned_2024_Base_dataset.xlsx")
```

```python
from sqlalchemy import create_engine

# Add a new column 'Data' to indicate the source of the data
base_2023['Data'] = '2023_Base'
base_2024['Data'] = '2024_Base'

# Find common columns between the two datasets
common_columns = list(set(base_2023.columns).intersection(base_2024.columns))

# Append the 2024_base data to the 2023_base based on common columns
merged_data = pd.concat([base_2023[common_columns], base_2024[common_columns]], ignore_index=True)

# Database connection details
db_username = 'postgres'
db_password = 'abc'
db_host = 'localhost'  # Or the server address
db_port = '5432'
db_name = 'postgres'

# Create a connection to PostgreSQL
engine = create_engine(f'postgresql://{db_username}:{db_password}@{db_host}:{db_port}/{db_name}')

# Define the table name
table_name = 'merged_2023_2024_base'

# Write the data to the PostgreSQL table
merged_data.to_sql(table_name, engine, if_exists='replace', index=False, chunksize=100000)
```
