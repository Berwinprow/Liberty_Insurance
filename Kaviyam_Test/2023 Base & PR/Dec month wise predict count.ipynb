```python
import pandas as pd

# Load your data into a DataFrame
data = pd.read_excel("Open Prediction with Format reasons and churned segment (S, N, W & E).xlsx") 

# Filter data for December 2024
filtered_data = data[
    (data['Policy End Date_YEAR'] == 2024) &
    (data['Policy End Date_MONTH'] == 12)
]

# Group by state and calculate counts for Yes and No labels
statewise_churn = filtered_data.groupby('state2').agg(
    Total_Policies=('Predicted Churn Label', 'count'),
    Yes_Count=('Predicted Churn Label', lambda x: (x == 'Yes').sum()),
    No_Count=('Predicted Churn Label', lambda x: (x == 'No').sum())
).reset_index()

# Save or display the results
statewise_churn.to_csv("Statewise_Churn_Dec_2024.csv", index=False)
print(statewise_churn)
```

```python
import pandas as pd

# Load your data into a DataFrame
data = pd.read_excel("Open Prediction with Format reasons and churned segment (S, N, W & E).xlsx") 

# Filter data for December 2024
filtered_data = data[
    (data['Policy End Date_YEAR'] == 2024) &
    (data['Policy End Date_MONTH'] == 12)
]

# Group by state and calculate counts for Yes and No labels
statewise_churn = filtered_data.groupby('state2').agg(
    Total_Policies=('Predicted Churn Label', 'count'),
    Yes_Count=('Predicted Churn Label', lambda x: (x == 'Yes').sum()),
    No_Count=('Predicted Churn Label', lambda x: (x == 'No').sum())
).reset_index()

# Calculate overall Yes and No counts
overall_counts = pd.DataFrame({
    'Category': ['Overall Yes Count', 'Overall No Count'],
    'Count': [
        filtered_data['Predicted Churn Label'].eq('Yes').sum(),
        filtered_data['Predicted Churn Label'].eq('No').sum()
    ]
})

# Save the data into an Excel file with two sheets
with pd.ExcelWriter("Statewise_Churn_Dec_2024.xlsx") as writer:
    statewise_churn.to_excel(writer, sheet_name="Statewise Data", index=False)
    overall_counts.to_excel(writer, sheet_name="Overall Counts", index=False)

print("Data saved successfully in Statewise_Churn_Dec_2024.xlsx")
```
