```python
import pandas as pd

# Load the Excel file
df = pd.read_excel('2023 PR.xlsx')

# Convert the 'Policy Issue Date' column to datetime format
df['Policy Issue date'] = pd.to_datetime(df['Policy Issue date'])

# Sort the dataframe by 'Policy Issue date' in descending order
df_sorted = df.sort_values(by='Policy Issue date', ascending=False)

# Identify duplicates to keep the first occurrence
duplicates_removed = df_sorted[df_sorted.duplicated(subset='Initial Policy No', keep='first')]

# Retain only the latest 'Policy Issue Date' for each 'Initial Policy No'
df_cleaned = df_sorted.drop_duplicates(subset='Initial Policy No', keep='first')

# Save the cleaned data to a new CSV file
df_cleaned.to_csv('cleaned_PR_dataset check.csv', index=False)

# Save the removed duplicate rows to a separate CSV file
duplicates_removed.to_csv('removed_duplicates PR file.csv', index=False)

print("Cleaned dataset and removed duplicates have been saved.")
```
