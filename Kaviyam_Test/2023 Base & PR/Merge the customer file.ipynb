```python
import pandas as pd

# Read the files
df1 = pd.read_excel("customer_aggregated_yearly_data (base data).xlsx")
df2 = pd.read_excel("customer_aggregated_yearly_data (RH).xlsx")
```

```python

# Step 2: Align the columns based on the order in df1
common_columns = list(df1.columns) + [col for col in df2.columns if col not in df1.columns]

# Reindex both DataFrames using the columns in the order of df1
df1 = df1.reindex(columns=common_columns)
df2 = df2.reindex(columns=common_columns)

# Step 3: Concatenate the DataFrames row-wise
merged_df = pd.concat([df1, df2], ignore_index=True)

# Step 4: Sort the merged data by 'CustomerID' and 'Start Year'
if 'CustomerID' in merged_df.columns and 'Start Year' in merged_df.columns:
    merged_df.sort_values(by=['CustomerID', 'Start Year'], inplace=True)

# Step 5: Save the result to a new Excel file
merged_df.to_csv("Merged_Customer_Data.csv", index=False)
```
