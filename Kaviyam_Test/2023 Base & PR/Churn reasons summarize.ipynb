```python
import pandas as pd
from transformers import pipeline

# Load the summarization pipeline
summarizer = pipeline("summarization")

 # Replace with your actual file path
data = pd.read_excel('D:\Liberty (Base data work)\Churn Reasons (Actual).xlsx')

# Ensure columns are named as expected
data.columns = [col.strip() for col in data.columns]  # Clean up column names if necessary

# Group by CustomerID and Start Year and summarize the Churn Reasons
summarized_data = []

# Function to summarize churn reasons
def summarize_text(text):
    if len(text) > 512:  # Transformer models have a token limit
        text = text[:512]
    summary = summarizer(text, max_length=50, min_length=25, do_sample=False)[0]['summary_text']
    return summary

# Process data group by group
for (customer_id, start_year), group in data.groupby(["CustomerID", "Start Year"]):
    churn_reasons = " ".join(group["Churn Reason"].dropna().tolist())
    summary = summarize_text(churn_reasons)
    summarized_data.append({"CustomerID": customer_id, "Start Year": start_year, "Churn Summary": summary})

# Convert summarized data to a DataFrame
summarized_df = pd.DataFrame(summarized_data)

# Save the summarized data to a new Excel file
output_file_path = "summarized_churn_reasons.xlsx"
summarized_df.to_excel(output_file_path, index=False)

print(f"Summarized data saved to {output_file_path}")
```
