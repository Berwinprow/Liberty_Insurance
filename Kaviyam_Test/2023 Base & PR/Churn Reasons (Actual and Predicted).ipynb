```python
import pandas as pd

df = pd.read_excel('D:\Liberty (Base data work)\Actual data (South).xlsx')
```

```python
def reason_for_churn(row):
    if row['Churn Label'] == 'Yes':
        reasons = []

        # Customer Tenure ≤ 1.50
        if row['Customer Tenure '] <= 1.50:
            reasons.append('Low Customer Tenure')

        # Average Discount (Overall) thresholds
        if row['Average Discount (Overall)'] <= 63.45:
            reasons.append('Low Discount Offered')
        elif row['Average Discount (Overall)'] > 72.03:
            reasons.append('High Discount Not Sustained')

        # Total Add-On Premium thresholds
        if row['Total Add on Premium (Overall)'] > 4469.07:
            reasons.append('High Add-On Premium')

        # No Claim Bonus (NCB) thresholds
        if row['Average NCB % Previous Year (Overall)'] <= 3.33:
            reasons.append('Low No Claim Bonus (NCB)')

        # Total Premium Payable (Overall) thresholds
        if row['Total Premium Payable (Overall)'] > 14141.0:
            reasons.append('High Total Premium Payable')

        # Number of Declines
        if row['Number of Declines'] > 0.5:
            reasons.append('Frequent Claim Declines')

        # Total OD Premium (Own Damage Premium)
        if row['Total OD Premium (Overall)'] > 6914.5:
            reasons.append('High Own-Damage Premium')
    
        # Total TP Premium Influence
        if row['Total TP Premium (Overall)'] > 8855.50:
            reasons.append('High TP Premium')

        # Number of Vehicles
        if row['Number of Vehicles'] <= 1.50:
            reasons.append('Low Engagement (Single Vehicle)')

        # Number of Claims > 0.50
        if row['Number of Claims'] > 0.50:
            reasons.append('Multiple Claims Filed')

        # Number of Policies ≤ 1.50
        if row['Number of Policies'] <= 1.50:
            reasons.append('Single Policy (Low Engagement)')

        return ', '.join(reasons) if reasons else 'Organic Churn'

    return ''

# Apply the function and export to CSV
df['Churn Reason'] = df.apply(reason_for_churn, axis=1)

df.to_csv('churn Reason_South Actual data.csv', index=False)

print(df.head())
```
