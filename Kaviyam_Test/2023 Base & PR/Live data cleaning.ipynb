```python
import pandas as pd
import re

df = pd.read_excel('Renewal Private Car data (November 2024).xlsx')

# Function to clean names
def clean_name(name):
    return re.sub(r'[^a-zA-Z0-9]', '', str(name)).lower()

# Create a new column with the cleaned INSUREDNAME
df['Cleaned_Insured name'] = df['Insured name '].apply(clean_name)

df['CustomerID_Base'] = (df['Cleaned_Insured name'].astype(str) + '_' +
                         df[' New Branch Name2'].astype(str))

# df['CustomerID'] = 'C' + (df.groupby('CustomerID_Base').ngroup() + 1).astype(str).str.zfill(5)

df.to_csv('cleaned_November 2024 data.csv', index=False)

print(df[['Insured name ', 'Cleaned_Insured name']].head())
```

```python
import pandas as pd

df = pd.read_excel('cleaned_October2024 data - Copy.xlsx')

# Initialize an empty list to store churn status labels
churned_labels = []

# Group the DataFrame by 'CustomerID'
for customer_id, group in df.groupby('CustomerID'):
    # Check if all policies for the current customer have the status 'Not Renewed'
    if (group['Policy Status'] == 'Not Renewed').all():
        churned_labels.append((customer_id, 'Yes'))  # Mark as churned ('Yes')
    else:
        churned_labels.append((customer_id, 'No'))   # Mark as not churned ('No')

# Convert the results into a DataFrame
churned_df = pd.DataFrame(churned_labels, columns=['CustomerID', 'Churn Label'])

# Merge the churn labels back into the original dataset
df = df.merge(churned_df, on='CustomerID', how='left')

df.to_csv("churned_customers.csv", index=False)
```
