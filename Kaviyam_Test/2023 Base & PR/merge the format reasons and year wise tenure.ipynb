```python
import pandas as pd

# Load the datasets
overall_data = pd.read_excel('Actaul Data with Reasons (S, N, W & E).xlsx')  # Replace with the actual file path
churn_data = pd.read_excel('Formatted_Churn_Reasons_Using_TextRank (Actual).xlsx')      # Replace with the actual file path

# Ensure the column names are consistent
# Replace 'Churn Reasons' and 'Formatted Churn Reason' with the actual column names in your datasets
churn_column = 'Churn Reason'
formatted_churn_column = 'Formatted Churn Reason'

# Merge the `Formatted Churn Reason` column to `overall_data` based on `Churn Reasons`
merged_data = pd.merge(
    overall_data,
    churn_data[[churn_column, formatted_churn_column]],
    how='left',
    on=churn_column
)

# Save the merged data to a new file
merged_data.to_csv('Actaul Data with Format Reasons (S, N, W & E).csv', index=False)
```

```python
import pandas as pd

# Load your dataset
# Replace 'your_data.csv' with the actual file path
data = pd.read_csv('Actaul Data with Format Reasons (S, N, W & E).csv')

# Convert 'Policy Start Date' and 'Policy End Date' to datetime
data['Policy Start Date'] = pd.to_datetime(data['Policy Start Date'])
data['Policy End Date'] = pd.to_datetime(data['Policy End Date'])

# Calculate tenure in months for each policy
data['Tenure (Months)'] = ((data['Policy End Date'].dt.year - data['Policy Start Date'].dt.year) * 12 +
                           (data['Policy End Date'].dt.month - data['Policy Start Date'].dt.month))

# Calculate cumulative tenure in months for each customer
data['Cumulative Tenure (Months)'] = data.groupby('CustomerID')['Tenure (Months)'].cumsum()

# Merge the tenure information back into the original data
final_data = data.copy()

# Save to CSV for review
final_data.to_csv('updated_Actaul Data with Format Reasons (S, N, W & E).csv', index=False)

# Display the final dataset
print(final_data)
```
