```python
import pandas as pd

df = pd.read_excel('Test Prediction (S, N, W & E).xlsx')
```

```python
# Calculate the match percentage between actual and predicted statuses
matches = df[df['Actual Churn Label'] == df['Predicted Churn Label']]
match_percentage = (len(matches) / len(df)) * 100
print(f"Match Percentage: {match_percentage:.2f}%")

mismatches = df[df['Actual Churn Label'] != df['Predicted Churn Label']]
mismatches.to_csv('mismatched_results (Overall test pred).csv', index=False)
```
