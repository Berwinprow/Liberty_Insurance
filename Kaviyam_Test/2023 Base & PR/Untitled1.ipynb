```python
import pandas as pd

# Load the main and renewed data into DataFrames
main_data = pd.read_excel("Pvt Car Jan'23 to  Dec 23 base Final.xlsx")  
renewed_data = pd.read_excel("cleaned_PR dataset.xlsx")  
```

```python
# Identify columns unique to renewed_data, plus ensure "New Policy No" is included
unique_columns = [col for col in renewed_data.columns if col not in main_data.columns] + ['New Policy No']

# Select only unique columns along with 'Initial Policy No' for matching
renewed_data_unique = renewed_data[['Initial Policy No'] + unique_columns].drop_duplicates()

# Initialize an empty list to store final rows with inserted matched data
final_rows = []

# Iterate through main_data rows
for _, main_row in main_data.iterrows():
    # Append the original row to final_rows
    final_rows.append(main_row)

    # Find the matching row in renewed_data_unique based on 'Initial Policy No'
    matching_row = renewed_data_unique[renewed_data_unique['Initial Policy No'] == main_row['Initial Policy No']]
    
    # If a matching row exists, append it right after the current main_row
    if not matching_row.empty:
        # Add the matched row (in dictionary format) to the final_rows list
        final_rows.append(matching_row.iloc[0])

# Convert the final rows back into a DataFrame
final_data = pd.DataFrame(final_rows)

# Save the final merged data to a new Excel file
final_data.to_excel("Merged_base_with_cleaned_PR_below_each_match.xlsx", index=False)

print("Data has been merged and saved with matched details inserted below each matched row, including 'New Policy No' and other unique columns from renewed_data.")
```
