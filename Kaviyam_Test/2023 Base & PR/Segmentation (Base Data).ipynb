```python
import pandas as pd

data = pd.read_excel('D:\Liberty (Base data work)\Open Prediction with reasons (S, N, W & E).xlsx')
```

```python
#Define thresholds for high/mid/low CLV and Payment Categories for all customers
high_average_discount = data['Average Discount (Overall)'].quantile(0.75)
mid_average_discount = data['Average Discount (Overall)'].quantile(0.5)
low_average_discount = data['Average Discount (Overall)'].quantile(0.25)

high_clv_threshold = data['CLV'].quantile(0.75)
mid_clv_threshold = data['CLV'].quantile(0.5)
low_clv_threshold = data['CLV'].quantile(0.25)

high_payment_threshold = data['Total Premium Payable (Overall)'].quantile(0.75)
mid_payment_threshold = data['Total Premium Payable (Overall)'].quantile(0.5)
low_payment_threshold = data['Total Premium Payable (Overall)'].quantile(0.25)

#Assign CLV and Payment categories
data['clv_category'] = data['CLV'].apply(
    lambda x: 'High' if x > high_clv_threshold else ('Mid' if x > mid_clv_threshold else 'Low')
)
data['payment_category'] = data['Total Premium Payable (Overall)'].apply(
    lambda x: 'High' if x > high_payment_threshold else ('Mid' if x > mid_payment_threshold else 'Low')
)
# Assign discount categories
data['discount_category'] = data['Average Discount (Overall)'].apply(
    lambda x: 'High' if x > high_average_discount else ('Mid' if x > mid_average_discount else 'Low')
)
#Determine churn status from a column indicating churn, assuming 'Not Renewed' indicates churn
data['churn_category'] = data['Predicted Churn Label'].apply(
    lambda x: 'High' if x == 'Yes' else 'Low'
)

# Segment Single Loan Customers using all possible combinations
def segment_customer(row):
    # Low Churn combinations
    if row['churn_category'] == 'Low':
        if row['payment_category'] == 'High':
            if row['discount_category'] == 'High':
                if row['clv_category'] == 'High':
                    return 'Core group (key investment)'
                elif row['clv_category'] == 'Mid':
                    return 'Core group (key investment)'
                elif row['clv_category'] == 'Low':
                    return 'Core group (key investment)'
            elif row['discount_category'] == 'Mid':
                if row['clv_category'] in ['High', 'Mid']:
                    return 'Core group (key investment)'
                elif row['clv_category'] == 'Low':
                    return 'Periphery group (selective investment)'
            elif row['discount_category'] == 'Low':
                if row['clv_category'] in ['High', 'Mid']:
                    return 'Periphery group (selective investment)'
                elif row['clv_category'] == 'Low':
                    return 'Periphery group (selective investment)'
        elif row['payment_category'] == 'Mid':
            if row['discount_category'] == 'High':
                if row['clv_category'] == 'High':
                    return 'Core group (key investment)'
                elif row['clv_category'] == 'Mid':
                    return 'Exploring ways to increase their value'
                elif row['clv_category'] == 'Low':
                    return 'Exploring ways to increase their value'
            elif row['discount_category'] == 'Mid':
                if row['clv_category'] in ['Mid', 'Low']:
                    return 'Exploring ways to increase their value'
                elif row['clv_category'] == 'High':
                    return 'Core group (key investment)'
            elif row['discount_category'] == 'Low':
                if row['clv_category'] in ['Low', 'Mid']:
                    return 'Exploring ways to increase their value'
                elif row['clv_category'] == 'High':
                    return 'Core group (key investment)'
        elif row['payment_category'] == 'Low':
            if row['discount_category'] == 'High':
                if row['clv_category'] == 'High':
                    return 'Core group (key investment)'
                elif row['clv_category'] in ['Mid', 'Low']:
                    return 'Exploring ways to increase their value'
            elif row['discount_category'] in ['Mid', 'Low']:
                if row['clv_category'] in ['Low', 'Mid']:
                    return 'Exploring ways to increase their value'
                elif row['clv_category'] == 'High':
                    return 'Core group (key investment)'

    # High Churn combinations
    elif row['churn_category'] == 'High':
        if row['payment_category'] == 'High':
            if row['discount_category'] == 'High':
                if row['clv_category'] == 'High':
                    return 'High revenue potential and Risk of leaving'
                elif row['clv_category'] == 'Mid':
                    return 'High Risking Customers'
                elif row['clv_category'] == 'Low':
                    return 'High Risking Customers'
            elif row['discount_category'] == 'Mid':
                if row['clv_category'] in ['High', 'Mid']:
                    return 'High revenue potential and Risk of leaving'
                elif row['clv_category'] == 'Low':
                    return 'High Risking Customers'
            elif row['discount_category'] == 'Low':
                if row['clv_category'] == 'High':
                    return 'High Risking Customers'
                elif row['clv_category'] in ['Mid', 'Low']:
                    return 'High Risking Customers'
        elif row['payment_category'] == 'Mid':
            if row['discount_category'] == 'High':
                if row['clv_category'] == 'High':
                    return 'High revenue potential and Risk of leaving'
                elif row['clv_category'] in ['Mid', 'Low']:
                    return 'High Risking Customers'
            elif row['discount_category'] == 'Mid':
                if row['clv_category'] in ['Mid', 'Low']:
                    return 'High Risking Customers'
                elif row['clv_category'] == 'High':
                    return 'High revenue potential and Risk of leaving'
            elif row['discount_category'] == 'Low':
                if row['clv_category'] in ['Low', 'Mid']:
                    return 'High Risking Customers'
                elif row['clv_category'] == 'High':
                    return 'High Risking Customers'
        elif row['payment_category'] == 'Low':
            if row['discount_category'] in ['High', 'Mid', 'Low']:
                if row['clv_category'] in ['High', 'Mid', 'Low']:
                    return 'High Risking Customers'

    # Default fallback
    return 'Other'


# Apply the segmentation function to the dataset
data['customer_segment'] = data.apply(segment_customer, axis=1)

# Save the results to a CSV file
data.to_csv('Open Prediction with reasons and segment (S, N, W & E).csv', index=False)

# Apply the segmentation function to the dataset
data['customer_segment'] = data.apply(segment_customer, axis=1)

data.to_csv('Open Prediction with reasons and segment (S, N, W & E).csv', index=False)
```
